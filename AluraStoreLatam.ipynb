### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Lista de dataframes
dataframes = [tienda, tienda2, tienda3, tienda4]

# New column names
new_columns = {
    'Unnamed: 0': 'id',
    'Producto': 'producto',
    'Categoría del Producto': 'categoria',
    'Precio': 'precio',
    'Costo de envío': 'costo_envio',
    'Fecha de Compra': 'fecha_compra',
    'Vendedor' : 'vendedor',
    'Lugar de Compra': 'lugar_compra',
    'Calificacón': 'calificacion',
    'Método de pago': 'metodo_pago',
    'Cantidad de cuotas': 'cantidad_cuotas',
    'lat': 'latitud',
    'lon': 'longitud'
}



# Rename columns for each dataframe
for df in dataframes:
    df.rename(columns=new_columns, inplace=True)

# Example: Display the head of the first dataframe to check the changes
tienda.head()
```

#1. Análisis de facturación



```python
#Análisis de facturación por tienda
total_revenues = []
net_revenues = []
store_names = []
# Iterar dataframes por tienda
for i, df in enumerate(dataframes):
    # Calcular facturación total
    df['revenue'] = df['precio']
    total_revenue = df['revenue'].sum()

    # Calcular costo de envío
    total_shipping_costs = df['costo_envio'].sum()

    # Calcular facturación neta
    net_revenue = total_revenue - total_shipping_costs

    # Imprimir los resultados de cada tienda
    print(f"Análisis de facturación para la tienda {i+1}:")
    print(f"Total de ingresos: {total_revenue}")
    print(f"Total de costos de envío: {total_shipping_costs}")
    print(f"Ingreso neto: {net_revenue}")
    print("-" * 20)

    total_revenues.append(total_revenue)
    net_revenues.append(net_revenue)
    store_names.append(f"Tienda {i+1}")

# Valor promedio facturado para cada tienda
for i, df in enumerate(dataframes):
    avg_order_value = df['revenue'].mean()
    print(f"Valor promedio del pedido para la tienda {i+1}: {avg_order_value}")


```

```python
# Análisis gráfico facturación total
plt.figure(figsize=(10, 6))
plt.bar(store_names, total_revenues, color='lightseagreen')
plt.xlabel("Tienda")
plt.ylabel("Ingresos totales")
plt.title("Ingresos totales por tienda")

# Incluir label
for i, v in enumerate(total_revenues):
    plt.text(i, v, f'{v/1000000:.1f}M', ha='center', va='bottom')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1000000:.1f}M'))
plt.show()


# Análisis gráfico facturación neta
plt.figure(figsize=(10, 6))
plt.bar(store_names, net_revenues, color='lightcoral')
plt.xlabel("Tienda")
plt.ylabel("Ingresos netos")
plt.title("Ingresos netos por tienda")

# Incluir labels
for i, v in enumerate(net_revenues):
    plt.text(i, v, f'{v/1000000:.1f}M', ha='center', va='bottom')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1000000:.1f}M'))
plt.show()
```

```python
# Análisis en una tabla

data = {'Tienda': store_names, 'Ingresos Totales': total_revenues, 'Ingresos Netos': net_revenues}
df = pd.DataFrame(data)

# calculo totales
total_revenue_sum = df['Ingresos Totales'].sum()
net_revenue_sum = df['Ingresos Netos'].sum()

# agregar totales a fila
totals_row = pd.DataFrame({'Tienda': ['Total'], 'Ingresos Totales': [total_revenue_sum], 'Ingresos Netos': [net_revenue_sum]})
df = pd.concat([df, totals_row], ignore_index=True)

# formato del dato moneda
pd.options.display.float_format = '${:,.2f}'.format

# Mostrar la tabla y darle estilo
styled_df = df.style.hide(axis='index')\
    .set_table_styles([
        {'selector': 'th', 'props': [('border', '1px solid black'), ('background-color', '#bbdefb '),('text-align', 'center')]},
        {'selector': 'td', 'props': [('border', '1px solid black'), ('text-align', 'center')]},
        {'selector': 'tr:nth-child(odd)', 'props': [('background-color', '#e3f2fd ')]},
        {'selector': 'tr:last-child', 'props': [('font-weight', 'bold'), ('background-color', '#bbdefb ')]},
            ])\
    .format({'Ingresos Totales': '${:,.2f}', 'Ingresos Netos': '${:,.2f}'}) # Apply the currency format here


styled_df
```

# 2. Ventas por categoría

```python
# Ventas por categoría
# Combinar df
all_stores_df = pd.concat([
    tienda.assign(store=1),
    tienda2.assign(store=2),
    tienda3.assign(store=3),
    tienda4.assign(store=4)
])

# calcular ventas por categoría para cada tienda
sales_by_category_store = all_stores_df.groupby(['categoria', 'store'])['revenue'].sum().reset_index()

# Pivotear la tabla c
pivot_table = sales_by_category_store.pivot_table(index='categoria', columns='store', values='revenue', fill_value=0)

# Agregar totales
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table.loc['Total'] = pivot_table.sum()
pivot_table = pivot_table.reset_index()

# Renombrar columnas
pivot_table.rename(columns={1: 'Tienda 1', 2: 'Tienda 2', 3: 'Tienda 3', 4: 'Tienda 4'}, inplace=True)

# Mostrar la tabla y darle estilo
styled_pivot_table = pivot_table.style.hide(axis='index')\
    .set_table_styles([
        {'selector': 'th', 'props': [('border', '1px solid black'), ('background-color', '#bbdefb '),('text-align', 'center')]},
        {'selector': 'td', 'props': [('border', '1px solid black'), ('text-align', 'center')]},
        {'selector': 'tr:nth-child(odd)', 'props': [('background-color', '#e3f2fd ')]},
        {'selector': 'tr:last-child', 'props': [('font-weight', 'bold'), ('background-color', '#bbdefb ')]},
            ])\
    .format({'Tienda 1': '${:,.2f}', 'Tienda 2': '${:,.2f}', 'Tienda 3': '${:,.2f}', 'Tienda 4': '${:,.2f}', 'Total': '${:,.2f}'})

styled_pivot_table
```

```python
# Ingresos netos por categoría
net_sales_by_category_store = all_stores_df.groupby(['categoria', 'store']).apply(lambda x: (x['revenue'] - x['costo_envio']).sum()).reset_index(name='net_sales')

net_sales_pivot_table = net_sales_by_category_store.pivot_table(index='categoria', columns='store', values='net_sales', fill_value=0)

net_sales_pivot_table['Total'] = net_sales_pivot_table.sum(axis=1)

net_sales_pivot_table.loc['Total'] = net_sales_pivot_table.sum()

net_sales_pivot_table = net_sales_pivot_table.reset_index()

net_sales_pivot_table.rename(columns={1: 'Tienda 1', 2: 'Tienda 2', 3: 'Tienda 3', 4: 'Tienda 4'}, inplace=True)

styled_net_sales_pivot_table = net_sales_pivot_table.style.hide(axis='index')\
    .set_table_styles([
        {'selector': 'th', 'props': [('border', '1px solid black'), ('background-color', '#bbdefb '),('text-align', 'center')]},
        {'selector': 'td', 'props': [('border', '1px solid black'), ('text-align', 'center')]},
        {'selector': 'tr:nth-child(odd)', 'props': [('background-color', '#e3f2fd ')]},
        {'selector': 'tr:last-child', 'props': [('font-weight', 'bold'), ('background-color', '#bbdefb ')]},
            ])\
    .format({col: '${:,.2f}' for col in net_sales_pivot_table.columns if col != 'categoria'}) # Apply currency format to all numeric columns

styled_net_sales_pivot_table
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos en la tienda
for i, df in enumerate(dataframes):
    print(f"\nAnálisis de productos más y menos vendidos para la tienda {i+1}:")

    # Contar apariciones productos
    product_counts = df['producto'].value_counts()

    # 5 productos más vendidos
    most_sold_products = product_counts.head(5)
    print("\nLos 5 productos más vendidos:")
    print(most_sold_products)

    # 5 productos menos vendidos
    least_sold_products = product_counts.tail(5)
    print("\nLos 5 productos menos vendidos:")
    print(least_sold_products)
    print("-" * 30)
```

```python
# Prodctos más y menos vendidos en gráficos

fig, axes = plt.subplots(2, 2, figsize=(20, 15))
axes = axes.flatten()

for i, df in enumerate(dataframes):

    product_counts = df['producto'].value_counts()

    most_sold_products = product_counts.head(5)

    least_sold_products = product_counts.tail(5)


    ax1 = axes[i]
    most_sold_products.plot(kind='barh', ax=ax1, color='skyblue')
    ax1.set_title(f'Tienda {i+1} - 5 Productos Más Vendidos')
    ax1.set_xlabel('Cantidad Vendida')
    ax1.set_ylabel('Producto')
    ax1.invert_yaxis()

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(20, 15))
axes = axes.flatten()

for i, df in enumerate(dataframes):

    product_counts = df['producto'].value_counts()


    least_sold_products = product_counts.tail(5)


    ax2 = axes[i]
    least_sold_products.plot(kind='barh', ax=ax2, color='lightcoral')
    ax2.set_title(f'Tienda {i+1} - 5 Productos Menos Vendidos')
    ax2.set_xlabel('Cantidad Vendida')
    ax2.set_ylabel('Producto')
    ax2.invert_yaxis()

plt.tight_layout()
plt.show()
```
